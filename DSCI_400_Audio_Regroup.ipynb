```
dmg = pandas.read_csv('/content/gdrive/MyDrive/DSCI400/Coswara-Data-master/Coswara-Data-master/combined_data.csv', index_col = 'id') # Upload (drag and drop) the csv file to your working directory
dmg
```

```
# regroup covid status
dmg.loc[(dmg['covid_status']=='healthy'),'status_regroup'] = 'healthy'
dmg.loc[(dmg['covid_status']=='no_resp_illness_exposed'),'status_regroup'] = 'exposed'
dmg.loc[(dmg['covid_status']=='positive_asymp'),'status_regroup'] = 'positive'
dmg.loc[(dmg['covid_status']=='positive_mild'),'status_regroup'] = 'positive'
dmg.loc[(dmg['covid_status']=='positive_moderate'),'status_regroup'] = 'positive'
dmg.loc[(dmg['covid_status']=='recovered_full'),'status_regroup'] = 'recovered'
dmg.loc[(dmg['covid_status']=='resp_illness_not_identified'),'status_regroup'] = 'resp_illness'

dmg.loc[(dmg['status_regroup']=='healthy'),'covid_pos'] = 0
dmg.loc[(dmg['status_regroup']=='recovered'),'covid_pos'] = 0
dmg.loc[(dmg['status_regroup']=='positive'),'covid_pos'] = 1
```

```
# split dataframe to covid positive vs. negative
dmg_pos = dmg.loc[(dmg['covid_pos']==1)]
dmg_neg = dmg.loc[(dmg['covid_pos']==0)]
```

```
# split train and test dataset
X_train_pos, X_test_pos, y_train_pos, y_test_pos = train_test_split(dmg_pos.iloc[:,:26], dmg_pos['covid_pos'], test_size=0.3, random_state=42)

X_train_neg, X_test_neg, y_train_neg, y_test_neg = train_test_split(dmg_neg.iloc[:,:26], dmg_neg['covid_pos'], test_size=0.3, random_state=42)
```

```
# get subject id for each set of subjects
train_pos_id = X_train_pos.index
test_pos_id = X_test_pos.index
train_neg_id = X_train_neg.index
test_neg_id = X_test_neg.index
```

```
# copy data to regrouped folders
cwd = os.getcwd()
for date in os.scandir("Extracted_data"):
  for subject in os.scandir(date):
    if subject.name in train_pos_id:
      %cp -av {cwd+"/Extracted_data/"+date.name+"/"+subject.name} {cwd+"/data_regroup/train/pos"}
    elif subject.name in train_neg_id:
      %cp -av {cwd+"/Extracted_data/"+date.name+"/"+subject.name} {cwd+"/data_regroup/train/neg"}
    elif subject.name in test_pos_id:
      %cp -av {cwd+"/Extracted_data/"+date.name+"/"+subject.name} {cwd+"/data_regroup/test/pos"}
    elif subject.name in test_neg_id:
      %cp -av {cwd+"/Extracted_data/"+date.name+"/"+subject.name} {cwd+"/data_regroup/test/neg"}
```

```
# print number of subjects in each group
print("train_pos",len(train_pos_id))
print("train_neg",len(train_neg_id))
print("test_pos",len(test_pos_id))
print("test_neg",len(test_neg_id))
```
